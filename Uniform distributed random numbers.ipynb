```python
import matplotlib.pyplot as plt


class Random:
    def __init__(self, Zn = 1):
        self.Zn = Zn
        self.a = 1664525
        self.b = 1013904223
        self.m = 2**32
    
    def random(self):
        self.Zn = (self.a * self.Zn + self.b) % self.m
        return self.Zn/self.m
    
def param_random(a, b, ran):
    return a + (b - a)*ran.random()
    
array_size = 100
ran = Random()

array =[param_random(0, 1, ran) for i in range(array_size)]

array_mean = 0

for i in range(array_size):
    array_mean = array_mean + array[i]

array_mean = array_mean/array_size

array_largest = array[0]
array_smallest = array[0]

for i in range(array_size - 1):
    if array[i] < array[i+1]:
        array_largest = array[i+1]
    if array[i] > array[i+1]:
        array_smallest = array[i+1]
    
array_range = array_largest - array_smallest

expected_mean = 0.5
expected_range = 1.0

estimated_a = array_mean - 0.5 * array_range
estimated_b = array_mean + 0.5 * array_range     

plt.hist(array, bins = 50, density = True)
plt.plot([0,1],[1,1], 'r-', lw=2, label = "Expected Uniform Distribution")
plt.legend()
plt.show()
```
