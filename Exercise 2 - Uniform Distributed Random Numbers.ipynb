```python
import matplotlib.pyplot as plt


class Random:
    def __init__(self, Zn = 788):
        self.Zn = Zn
        self.a = 1664525
        self.b = 1013904223
        self.m = 2**32
    
    def random(self):
        self.Zn = (self.a * self.Zn + self.b) % self.m
        return self.Zn/self.m  # Generate random no using linear congruential generator
    
def param_random(a, b, ran):
    return a + (b - a)*ran.random() #Parameterize the random no generator
    
array_size = 100
ran = Random()

array =[param_random(0, 1, ran) for i in range(array_size)] #Sample Set

array_mean = 0

for i in range(array_size):
    array_mean = array_mean + array[i]

array_mean = array_mean/array_size  #Estimate mean
    
array_range = max(array) - min(array) #Estimate range

print(array_mean)
print(array_range)

#expected_mean = 0.5
#expected_range = 1.0

estimated_a = array_mean - 0.5 * array_range
estimated_b = array_mean + 0.5 * array_range     

plt.hist(array, bins = 10, density = True)   #Plotting histogram to check the distribution
plt.plot([0,1],[1,1], 'r-', lw=2, label = "Expected Uniform Distribution")
plt.legend()
plt.show()
```
